Code
steps to capture a video:

- use cv2.VideoCapture() to get a video capture object for the camera.
- Video is a combination of frame
- when we capture live video, it is like a infinite frames till we stop the video
- set up a infinite wile loop and use the read() method to read the frames using the above created object.
- Use cv2.imshow() method to show the frames in the video
- Breaks the loop when the user clicks a specific key.

```python
cv2.imread() : Image read
cv2.imshow() : Image show
cv2.imwrite() : Image write
cv2.VideoCapture(0) # Here  will open the laptop camera capture our videos
```

```python
import cv2

# define a video capture object
vid = cv2.VideoCapture(0)

while True:
    # Capture video frame by frame
    ret, frame = vid.read()
    # ret = boolean value if ret = True it means Frame is capture, if its False frame is not capture
    print(ret)

    # Display the resulting frame
    cv2.imshow('frame', frame)

    # the q button is set as the quitting button you may use any other also
    # 1 means one milisecond
    # 1 and 0xFF its try identity an 8 digit key
    # ord('q') : ASCII 8 digits
    # so it will check you entered 8 digit key or not
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break 
# After the loop realeas the cap object 
vid.release()

# Destroy all windows
cv2.destroyAllWindows()
```

```python
#  importing all library
import cv2
import os

# read the video from specific path

cam = cv2.VideoCapture(r"C:\Users\sahil\Downloads\los_angeles.mp4")

try: 
#     # creating a folder name data
    if not os.path.exists('data'):
        os.makedirs('data')
# if not created then raise error

except OSError:
    print('Error :  creating directory of data')

# frame

currentframe = 0

while True:
    # reading from frame
    ret, frame = cam.read()
    if ret:
        # if video is still left continue creating images
        name = "./data/frame" + str(currentframe) + '.jpg'
        print("Creating..", name)

        #writing the extracted images
        cv2.imwrite(name, frame)
        # increasing counter so that it will
        # show how many frames are created
        currentframe +=1
    else:
        break

#after the loop release the cap object
cam.release()
# destroy all windows
cv2.destroyAllWindows()
```

**Write the text on the image**

```python
import cv2
vid = cv2.VideoCapture(0)

while True:
    ret, frame = vid.read()
    print(ret)

    cv2.imshow('frame', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
vid.release()
cv2.destroyAllWindows()
```

```python
import cv2
vid = cv2.VideoCapture(0)

while True:
    ret, frame = vid.read()

    # describe the font to be used
    font = cv2.FONT_HERSHEY_DUPLEX

    # use putText() method for 
    # inserting text on video  R=0,G=255,B=255
    cv2.putText(frame,
        'Hi Sahil',
        (50,50),# position of text
        font,
        1,# font size
        (0,25,155), # color of the font
        2,# thickness 
        cv2.LINE_4
    )

    cv2.imshow('frame', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
vid.release()
cv2.destroyAllWindows()
```
