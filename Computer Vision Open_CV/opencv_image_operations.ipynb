**Convert color image to gray  image**

```python
# read the image
image = cv2.imread(r"C:\Users\sahil\Downloads\actor.jpg")
# gray scale conversion
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow('ABD_BW', gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

**CROP The Image**
- In the image we can select a perticuler part of image
- Image has pixels of rows and columns
- slice around rows slice around the column values of image

```python
cropped_image = image[150:550, 350:640]
plt.imshow(cropped_image)
```
