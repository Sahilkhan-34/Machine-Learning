```python
categorical_cols = df.select_dtypes(include='object').columns.tolist()

# Create a figure with subplot
no_cols = len(categorical_cols)
ncol = 3
nrow = (no_cols + ncol) // ncol
fig, axs = plt.subplots(nrows=nrow, ncols=ncol, figsize=(15, 5*nrow))
axs = axs.flatten()


# Create Countplot for top 5 values of each categorical variable
for i, var in enumerate(categorical_cols):
    # here we calculate top 5 values of each categorical columns
    top_5_values = df[var].value_counts().nlargest(5).index

    # here we filter the top 5 values put in the data frame name filter_values_df 
    filter_values_df = df[df[var].isin(top_5_values)]

    # plot countplot
    sns.countplot(x=var, data=filter_values_df, ax=axs[i])
    axs[i].set_title(var) # title of plot
    axs[i].tick_params(axis='x', rotation=90) # rotate name in x axis 90 degree to readiablity

# Remove any extra empty subplots if needed
if no_cols < len(axs):
    for i in range(no_cols, len(axs)):
        fig.delaxes(axs[i])

#Adjusting spacing between subplot
fig.tight_layout()

# show plots
plt.show()
```

```python
# Plotting count plot with hue Customer Status

# remove target Column
if 'Customer Status' in categorical_cols:
    categorical_cols.remove('Customer Status')

# create figure and sub plot

num_cols = len(categorical_cols)
ncol = 3
nrow = (num_cols + ncol) // ncol
fig, axs = plt.subplots(nrows=nrow, ncols=ncol, figsize=(15, 5*nrow))
axs = axs.flatten()


for i, var in enumerate(categorical_cols):
    top_5_values = df[var].value_counts().nlargest(5).index
    filter_values_df = df[df[var].isin(top_5_values)]
    sns.countplot(x=var, hue='Customer Status', data=filter_values_df, ax=axs[i])
    axs[i].set_xticklabels(axs[i].get_xticklabels(), rotation=90)


# removing extra subplots
if num_cols < len(axs):
    for i in range(num_cols, len(axs)):
        fig.delaxes(axs[i])

# Adjust the space between subplots
fig.tight_layout()

# Show the plot
plt.show()
```
